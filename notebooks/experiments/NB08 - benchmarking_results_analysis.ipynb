```python
import pandas as pd 
```

```python
df = pd.read_excel("../../data/ChatLSE Model Benchmarking.xlsx", header=1, usecols='B:N')

df.columns = ["tester_name", "model", "is_summariser", "question", "accuracy", "coherence", "fluency", "groundedness", "relevance", "full_json_log", "conversation_length", "comments", "total_score"]
df_scores = df[["model", "question", "accuracy", "coherence", "fluency", "groundedness", "relevance", "total_score"]].dropna()

df_scores["model"] = df_scores["model"].apply(lambda x: x.rstrip())
df_scores["model"] = df_scores["model"].apply(lambda x: x.lower())
df_scores["question"] = df_scores["question"].apply(lambda x: x.lower())

df_scores.head()
```

```python
df_grouped_qn = df_scores.groupby(["question", "model"]).agg("mean").round(2)

df_grouped_qn
```
